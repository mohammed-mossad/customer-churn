```python
import pandas as pd
import numpy as np
df = pd.read_csv(r'cleaned_credit_card_customers.csv')

df.head()
```

```python
customer_count = df['customer_number'].count()
print(f"The number of customer numbers is: {customer_count}")

attrition_counts = df.groupby('attrition_flag')['attrition_flag'].count()
attrition_counts
```

```python
from sklearn.preprocessing import LabelEncoder

lable_encoder = LabelEncoder()

df['gender'] = lable_encoder.fit_transform(df['gender'])
df['attrition_flag'] = lable_encoder.fit_transform(df['attrition_flag'])
```

```python
df.head()
```

```python
from sklearn.preprocessing import OneHotEncoder
ohe = OneHotEncoder()

ohe.fit_transform(df[['education_level', 'marital_status', 'income_category']]).toarray()
```

```python
feature_array = ohe.fit_transform(df[['education_level', 'marital_status', 'income_category']]).toarray()

ohe.categories_
```

```python
feature_lables = ohe.categories_
```

```python
feature_labels_flat = np.concatenate([arr.ravel() for arr in feature_lables])
print(feature_labels_flat)
```

```python
pd.DataFrame(feature_array, columns=feature_labels_flat)
```

```python
features = pd.DataFrame(feature_array, columns=feature_labels_flat)
print(features)
```

```python
df2 = pd.concat([df, features], axis=1)
```

```python
df2.head()
```

```python
df2.dtypes
```

```python
df_new = df.drop(columns=['date_of_birth', 'card_category', 'education_level', 'marital_status', 'income_category'])
```

```python
df_new.head()
```

```python
df_new.dtypes
```

```python
df3 = pd.concat([df_new, features], axis=1)
```

```python
df3.head()
```

```python
df3.dtypes
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report, accuracy_score

X = df3.drop(columns=['attrition_flag'])
y = df3['attrition_flag']

# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Initialize the RandomForestClassifier
rfc = RandomForestClassifier(random_state=42)

# Train the model
rfc.fit(X_train, y_train)

# Make predictions
y_pred = rfc.predict(X_test)

# Evaluate the model
accuracy = accuracy_score(y_test, y_pred)
classification_rep = classification_report(y_test, y_pred)

# Print evaluation results
print(f'Accuracy: {accuracy:.4f}')
print('Classification Report:')
print(classification_rep)
```
